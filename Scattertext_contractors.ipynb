```python
"""
Import Statements
"""

# Base
from collections import Counter
import re

import pandas as pd

# Plotting
import squarify
import matplotlib.pyplot as plt
import seaborn as sns

# NLP Libraries
import spacy
from spacy.tokenizer import Tokenizer
from nltk.stem import PorterStemmer

nlp = en_core_web_md.load()
tokenizer = Tokenizer(nlp.vocab)#breaks text, it creates character indexes instead of splitting words up

STOP_WORDS = nlp.Defaults.stop_words.union(['him', 'deaf', 'her','told','told him','ins', '1 check','I', 'i"m', 'i', ' ', 'it', "it's", 'it.','they','coffee','place','they', 'the', 'this','its', 'l','-','they','this','don"t','the ', ' the', 'it', 'i"ve', 'i"m'])

STOP_WORDS
```

```python
tokens = []

for doc in tokenizer.pipe(df['text'], batch_size=800):
    
    doc_tokens = []
    
    for token in doc: 
        if token.text not in STOP_WORDS:
            doc_tokens.append(token.text.lower())   
    tokens.append(doc_tokens)
    
fast_biz['tokens'] = tokens
```

```python
def count(docs):

        word_counts = Counter()
        appears_in = Counter()
        
        total_docs = len(docs)

        for doc in docs:
            word_counts.update(doc)
            appears_in.update(set(doc))

        temp = zip(word_counts.keys(), word_counts.values())
        
        wc = pd.DataFrame(temp, columns = ['word', 'count'])

        wc['rank'] = wc['count'].rank(method='first', ascending=False)
        total = wc['count'].sum()

        wc['pct_total'] = wc['count'].apply(lambda x: x / total)
        
        wc = wc.sort_values(by='rank')
        wc['cul_pct_total'] = wc['pct_total'].cumsum()

        t2 = zip(appears_in.keys(), appears_in.values())
        ac = pd.DataFrame(t2, columns=['word', 'appears_in'])
        wc = ac.merge(wc, on='word')

        wc['appears_in_pct'] = wc['appears_in'].apply(lambda x: x / total_docs)
        
        return wc.sort_values(by='rank')

wc = count(fast_biz['tokens'])
wc_top20 = wc[wc['rank'] <= 20]

squarify.plot(sizes=wc_top20['pct_total'], label=wc_top20['word'], alpha=.8 )
plt.axis('off')
plt.show()
```

```python
# extract pronouns 
def get_lemmas(text):

    lemmas = []
    
    doc = nlp(text)
    
    # Something goes here :P
    for token in doc:
        if ((token.is_stop == False) and (token.is_punct == False)) and (token.pos_!= 'PRON') and (token not in STOP_WORDS):#exclude pronounds
            lemmas.append(token.lemma_)
    
    return lemmas
fast_biz['lemmas'] = fast_biz['text'].apply(get_lemmas)
```

```python
wc = count(fast_biz['lemmas'])
wc_top20 = wc[wc['rank'] <= 20]

squarify.plot(sizes=wc_top20['pct_total'], label=wc_top20['word'], alpha=.8 )
plt.axis('off')
plt.show()
```

```python
fast_biz['lemmas'] = [[word.lower() for word in text] for text in fast_biz['lemmas']]
```

```python
fast_biz['lemmastr'] = fast_biz['lemmas'].str.join(' ')
fast_biz.head(2)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

data= fast_biz['lemmastr'].values

vectorizer = CountVectorizer(stop_words='english')

# tokenize and build vocab
vectorizer.fit_transform(data)


# Create a Vocabulary
# The vocabulary establishes all of the possible words that we might use.
vectorizer.vocabulary_

# The vocabulary dictionary does not represent the counts of words!!
dtm = vectorizer.transform(data)
dtm.todense() 
```

```python
import scattertext as st
import spacy
from pprint import pprint

# convention_df = st.SampleCorpora.ConventionData2012.get_data()  
# convention_df.iloc[0]
nlp = spacy.load("en_core_web_md")
corpus = st.CorpusFromPandas(fast_biz, 
                             category_col='review_stars', 
                             text_col='text',
                             nlp=nlp).build()
```

```python
pprint(list(corpus.get_scaled_f_scores_vs_background().index[:10]))
```

```python
term_freq_df = corpus.get_term_freq_df()
term_freq_df['highratingscore'] = corpus.get_scaled_f_scores('5')
pprint(list(term_freq_df.sort_values(by='highratingscore', ascending=False).index[:10]))
```
